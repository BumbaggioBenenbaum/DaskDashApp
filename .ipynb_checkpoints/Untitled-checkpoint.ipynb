```python
from dask.distributed import get_worker
import pandas as pd
```

```python
def lemmatize(
        partition:pd.DataFrame,
        inp:str
    ) -> pd.DataFrame:

        worker = get_worker()
        try:
            nlp = worker.nlp
        except AttributeError:
            nlp = spacy.load(config.spacy.model)
            worker.nlp = nlp

        return pd.Series(nlp.pipe(partition[inp].values))
```
