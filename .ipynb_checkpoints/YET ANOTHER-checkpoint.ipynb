```python
def token_criteria(
    _token: spacy.tokens.token.Token,
    stopwords: Union[set, list] = ["ale"]
) -> bool:
    token = str(_token)
    return (
        2 < len(token) < 20 
        and token.isalnum() 
        and token not in stopwords 
        and str(_token.lemma_) not in stopwords
    )


def clean_sents(
    text: spacy.tokens.doc.Doc
) -> Tuple[list, list]:
    _sents = list(text.sents)
    raw_sents = [str(x) for x in _sents]
    sents = [[str(tok.lemma_) for tok in sent if token_criteria(tok)] for sent in _sents]
    return raw_sents, sents


def dict_from_varnames(
     _dict:dict,
     names: list
)-> Dict[str, Any]:
    return {
        name: _dict.get(name)
        for name in names
        if name in _dict
    }  


def find_hashtags(text:str)-> List[str]:
    return re.findall(r"#(\w+)", text)
    
    
def transform_text(
    text: spacy.tokens.doc.Doc
) -> Dict[str, Any]:

    _sents = clean_sents(text)
    sents = _sents[0]
    lemmatized_sent_lists = _sents[1]
    lemmatized_sent_strings = [" ".join(sent) for sent in lemmatized_sent_lists]
    lemmata = list(chain.from_iterable(lemmatized_sent_lists))
    unique_lemmata = list(set(lemmata))
    lemmata_bigrams = list(nltk.bigrams(lemmata))
    lemmata_trigrams = list(nltk.trigrams(lemmata))
    text = str(text)
    hashtags = find_hashtags(text)
    
    return dict_from_varnames(
        locals(),
        [x for x in locals() if not x.startswith("_")]
    )


def dict_to_df(d:dict) -> pd.DataFrame:
    data = d.items()
    index = d.keys()
    return pd.DataFrame(data, index = index).transpose().tail(1)
```

```python
class MetaStrEnum(StrEnum):
    
    def __init_subclass__(cls):
        cls.as_meta = lambda: sorted([(x, None) for x in cls.__dict__.get("_member_names_", [])])
        return cls

    
class PreProcessingColumns(MetaStrEnum):
    text = auto()
    

class PostProcessingColumns(MetaStrEnum):
    lemmata = auto()
    lemmata_bigrams = auto()
    lemmata_trigrams = auto()
    lemmatized_sent_lists = auto()
    lemmatized_sent_strings = auto()
    sents = auto()
    unique_lemmata = auto()
    hashtags = auto()
    text = auto()
    
    
def process_text(partition:pd.DataFrame) -> pd.DataFrame:
    worker = get_worker()
    try:
        nlp = worker.nlp
    except AttributeError:
        nlp = spacy.load('pl_core_news_sm', disable=['ner'])
        worker.nlp = nlp
        
    res = []
    for doc in nlp.pipe(partition):
        res.append(dict_to_df(transform_text(doc)))    
    concated_new = pd.concat(res)
    concated_new.index = partition.index
    concated_new = concated_new[sorted(concated_new.columns)]
    
    
    #HERE COUNT BI/TRIGRAMS!
    
    
    return concated_new


def export_grams(df, alias) -> None:
    c = gather_counters(df[f"lemmata_{alias}grams"].map_partitions(lambda p: count_ngrams(p), meta=(None, str)))
    df = pd.DataFrame(c.most_common(500))
    df.columns = ["words", "count"]
    df.to_csv(f"{alias}grams.csv", index=False)

def count_ngrams(partition:pd.DataFrame) -> Counter:
    return Counter([x for y in partition for x in y])


def gather_counters(counter_series:pd.Series) -> Counter:
    c = Counter()
    for counter in counter_series:
        c += counter
    return c

```
